```python
import numpy as np
import pandas as pd
import scipy
from scipy.stats import wilcoxon
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
def read_result(path = './results/1SRC_noadapt_', metric='mae'):
    path = path+metric+'.csv'
    df = pd.read_csv(path, header=0, index_col=0)
    if metric=='mae':
        Z = np.array([[float(df.values[i,j][7:13]) for j in range(df.shape[1])] for i in range(df.shape[0])])
        df = pd.DataFrame(data = Z, columns  = df.columns)
    return df

metric = 'mae'
df_no = read_result(path = './results/1SRC_noadapt_', metric=metric)
df_1srcadisc = read_result('./results/1SRC_Adisc_', metric=metric)
df_dann = read_result('./results/1SRC_DANN_', metric=metric)
df_mdan = read_result('./results/MDAN_', metric=metric)
df_msdaadisc = pd.read_csv('./results/ADisc_MSDA_mae.csv', index_col=0).T
df_tot = pd.concat((df_no.mean(axis=0), df_1srcadisc.mean(axis=0), df_dann.mean(axis=0), df_mdan.mean(axis=0), df_msdaadisc.mean(axis=1)),axis=1)
df_tot.columns = ['No-adapt', 'ADisc', 'DANN', 'MDAN', 'ADisc-MSDA']
#df_tot['best'] = df_tot.columns[df_tot.values.argmin(axis=1)]
display(df_tot)
```
